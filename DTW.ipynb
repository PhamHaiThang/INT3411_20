```python
a1 = "/content/A/a1.wav"
a2 = "/content/A/a2.wav"
a3 = "/content/A/a3.wav"

b1 = "/content/B/b1.wav"
b2 = "/content/B/b2.wav"
b3 = "/content/B/b3.wav"

ban1 = "/content/Ban/ban1.wav"
ban2 = "/content/Ban/ban2.wav"
ban3 = "/content/Ban/ban3.wav"

len1 = "/content/Len/len1.wav"
len2 = "/content/Len/len2.wav"
len3 = "/content/Len/len3.wav"

nhay1 = "/content/Nhay/nhay1.wav"
nhay2 = "/content/Nhay/nhay2.wav"
nhay3 = "/content/Nhay/nhay3.wav"

phai1 = "/content/Phai/phai1.wav"
phai2 = "/content/Phai/phai2.wav"
phai3 = "/content/Phai/phai3.wav"

trai1 = "/content/Trai/trai1.wav"
trai2 = "/content/Trai/trai2.wav"
trai3 = "/content/Trai/trai3.wav"

xuong1 = "/content/Xuong/xuong1.wav"
xuong2 = "/content/Xuong/xuong2.wav"
xuong3 = "/content/Xuong/xuong3.wav"
```

```python
def normalize_mfcc(mfcc):
    for i in range(mfcc.shape[1]):
        mfcc[:,i] = mfcc[:,i] - np.mean(mfcc[:,i])
        mfcc[:,i] = mfcc[:,i] / np.max(np.abs(mfcc[:,i]))
    return mfcc
```

**DTW word "a"**

```python
ipd.Audio(a1)
```

```python
ipd.Audio(a2)
```

```python
ipd.Audio(a3)
```

```python
lib_a1 = librosa.load(a1)[0]
lib_a2 = librosa.load(a2)[0]
lib_a3 = librosa.load(a3)[0]
```

```python
template_a1 = normalize_mfcc(librosa.feature.mfcc(y=lib_a1))
template_a2 = normalize_mfcc(librosa.feature.mfcc(y=lib_a2))
template_a3 = normalize_mfcc(librosa.feature.mfcc(y=lib_a3))
```

```python
metric = 'euclidean'
```

```python
metric = 'cosine'
```

```python
cost01_a, align01_a = librosa.sequence.dtw(template_a1, template_a2, metric=metric)
cost02_a, align02_a = librosa.sequence.dtw(template_a1, template_a3, metric=metric)
```

```python
count_a = np.ones(len(template_a1[0]))

summ_a = template_a1.copy()
```

```python
for pair_a in align01_a:
    t, q = pair_a
    count_a[t] += 1
    summ_a[:,t] += template_a2[:,q]
```

```python
for pair_a in align02_a:
    t, q = pair_a
    count_a[t] += 1
    summ_a[:,t] += template_a3[:,q]
```

```python
average_a = summ_a / count_a
average_a.shape
```

```python
average_template_a = average_a
```

```python
librosa.sequence.dtw(normalize_mfcc(librosa.feature.mfcc(y=librosa.load('/content/A/a1.wav')[0])), average_template_a, metric=metric)[0][-1,-1]
```

```python
librosa.sequence.dtw(normalize_mfcc(librosa.feature.mfcc(y=librosa.load('/content/A/a2.wav')[0])), average_template_a, metric=metric)[0][-1,-1]
```

```python
librosa.sequence.dtw(normalize_mfcc(librosa.feature.mfcc(y=librosa.load('/content/A/a3.wav')[0])), average_template_a, metric=metric)[0][-1,-1]
```
