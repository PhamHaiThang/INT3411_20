```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

**Loading audio file**


```python
audio_file = "/c53.wav" 
```

```python
ipd.Audio(audio_file)
```

```python
signal, sr = librosa.load(audio_file)
```

```python
signal.shape
```

**Extracting MFCCs**

```python
mfccs = librosa.feature.mfcc(signal, n_mfcc=13, sr=sr)
```

```python
mfccs.shape
```

**Visualize MFCCs**

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(mfccs, x_axis="time", sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

**Calculating delta and deltadelta MFCCs**

```python
delta_mfccs = librosa.feature.delta(mfccs)
delta2_mfccs = librosa.feature.delta(mfccs, order=2)
```

```python
delta_mfccs.shape
```

```python
mfccs.shape
```

**Visualize Delta and Deltadelta MFCCs**

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(delta_mfccs, x_axis="time", sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(delta2_mfccs, x_axis="time", sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
comprehensive_mfccs = np.concatenate((mfccs, delta_mfccs, delta2_mfccs))
```
